# **Monochrome image (Binary Image)**

```python
import numpy as np

# Create a monochrome image
img = np.zeros((256, 256), dtype=np.uint8)
img[100:200, 100:200] = 255  # Set a white square

import matplotlib.pyplot as plt
plt.imshow(img, cmap='gray')
plt.show()
```

# **Halftone Image**

```python
import numpy as np

# Create a halftone image
img = np.random.randint(0, 256, size=(256, 256), dtype=np.uint8)
halftone_img = np.zeros((256, 256), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        if img[i, j] > 128:
            halftone_img[i, j] = 255
        else:
            halftone_img[i, j] = 0

import matplotlib.pyplot as plt
plt.imshow(halftone_img, cmap='gray')
plt.show()
```

# **Convert Color to Greyscale Image**

```python
import numpy as np
# Load the color image
img = np.random.randint(0, 256, size=(256, 256, 3), dtype=np.uint8)
# Convert the image to grayscale
gray_img = np.zeros((256, 256), dtype=np.uint8)
for i in range(256):
    for j in range(256):
        # Calculate the grayscale value using the formula: gray = (R + G + B) / 3
        gray_img[i, j] = (img[i, j, 0] + img[i, j, 1] + img[i, j, 2]) // 3
# Display the grayscale image
import matplotlib.pyplot as plt
plt.imshow(gray_img, cmap='gray')
plt.show()
```

# **Convert GreyScale Image to Color Image**

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
gray_img = np.random.randint(0, 256, size=(256, 256), dtype=np.uint8)

# Convert the grayscale image to a color image
color_img = np.zeros((256, 256, 3), dtype=np.uint8)
for i in range(256):
    for j in range(256):
        # Add a constant color to each pixel
        color_img[i, j, 0] = gray_img[i, j]  # Red
        color_img[i, j, 1] = gray_img[i, j]  # Green
        color_img[i, j, 2] = gray_img[i, j]  # Blue

# Display the color image
plt.imshow(color_img)
plt.show()
```

**If you want to add a specific color to the image, you can modify the code to add a constant value to each color channel. For example, to add a red tint to the image**:

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
gray_img = np.random.randint(0, 256, size=(256, 256), dtype=np.uint8)

# Convert the grayscale image to a color image
color_img = np.zeros((256, 256, 3), dtype=np.uint8)
for i in range(256):
    for j in range(256):
        # Add a red tint to each pixel
        color_img[i, j, 0] = gray_img[i, j]  # Red
        color_img[i, j, 1] = int(gray_img[i, j] * 0.7)  # Green
        color_img[i, j, 2] = int(gray_img[i, j] * 0.5)  # Blue

# Display the color image
plt.imshow(color_img)
plt.show()
```
